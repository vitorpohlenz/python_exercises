````python
def linear_func(x, a, b):
    return a * x + b

# Generate noisy data
np.random.seed(SEED)

x_data = np.linspace(0, 10, 100)
y_data = 2 * x_data + 3 + np.random.normal(0, 2.5, 100)

# Fit the data
params, _ = curve_fit(linear_func, x_data, y_data)
a, b = params

# Plot original data and fitted curve
plt.scatter(x_data, y_data, label='Data')
plt.plot(x_data, linear_func(x_data, a, b), color='blue', label='Fitted Line')
plt.legend()
plt.show()
````

````python
# Analytical solution is: y = e^(-2x)
def ordinary_diff_equation(y, x):
    return -2 * y

x = np.linspace(0, 5, 100)
y0 = 1
y = odeint(ordinary_diff_equation, y0, x)
plt.plot(x, y)
plt.title("Solution of dy/dx = -2y")
plt.show()
````

````python
# Number of samplepoints
N = 100
# Space between samples or "delta time" between samples
T = 1.0 / 800.0

# Generate the time vector
x = np.linspace(0.0, N*T, N)

# Generate the signal
y = np.sin(50.0 * 2*np.pi*x)

# FFT of signal
yfft = fft(y)

# Taking just the unilateral spectrum (associated with Real part)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)
yf = 2.0/N * np.abs(yfft[:N//2])

fig, ax = plt.subplots()
ax.plot(xf, yf)
plt.show()
````

````python
# Generating the data
x = np.linspace(0, 10, 5)
# Using sin(x) as ground truth
y = np.sin(x)

f_interp = interp1d(x, y, kind='cubic')
x_new = np.linspace(0, 10, 50)
y_new = f_interp(x_new)

plt.plot(x, y, 'o', label='Original Data')
plt.plot(x_new, y_new, '-', label='Interpolated Data')
plt.legend()
plt.show()
````

````python
# Linear coeficients
A = np.array([[2, 1], [1, -1]])
# Constants
b = np.array([3, 1])
solution = linalg.solve(A, b)
print(f"Solution to the linear system is [x y] = {solution}")

# Plotting the solution
fig = plt.figure()
plt.xlabel('x')
plt.ylabel('y')
ax = fig.gca()
ax.plot(solution[0], solution[1], 'ro')

plt.grid()
plt.show()
````
