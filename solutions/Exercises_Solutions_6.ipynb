````python
iris_data = px.data.iris()

# Plot
fig = px.scatter(
    data_frame = iris_data,
    x='petal_length',
    y='petal_width',
    color='species')


# Vertical Line
fig.add_vline(x=iris_data['petal_length'].mean(), line_dash="dash", line_color="yellowgreen")

# Horizontal Line
fig.add_hline(y=iris_data['petal_width'].mean(), line_dash="dash", line_color="pink")

fig.show()
````

````python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])
fig.update_layout(title="3D Surface Plot of sin(x) * cos(y)", autosize=False,
                  width=700, height=700)
fig.show()
````
